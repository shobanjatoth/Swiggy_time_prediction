# Encoding

```python
def binary_encode(df, binary_map):
    """
    binary_map: {column_name: {old_value: new_value}}
    """
    df = df.copy()
    
    for col, mapping in binary_map.items():
        df[col] = df[col].map(mapping)
        
    return df
```

```python
def ordinal_encode(df, ordinal_map):
    """
    ordinal_map: {column_name: [ordered_categories]}
    """
    df = df.copy()
    
    cols = list(ordinal_map.keys())
    categories = list(ordinal_map.values())
    
    encoder = OrdinalEncoder(categories=categories)
    df[cols] = encoder.fit_transform(df[cols])
    
    return df
```

```python
def nominal_encode(df, nominal_cols, drop_first=True):
    """
    drop_first=True avoids dummy variable trap (recommended for linear models)
    """
    df = df.copy()
    
    df = pd.get_dummies(
        df,
        columns=nominal_cols,
        drop_first=drop_first
    )
    
    return df
```

```python
# Binary
binary_map = {
    'festival': {'no': 0, 'yes': 1}
}

# Ordinal
ordinal_map = {
    'traffic': ['low', 'medium', 'high', 'jam'],
    'order_day_of_week': [
        'monday', 'tuesday', 'wednesday',
        'thursday', 'friday', 'saturday', 'sunday'
    ],
    'order_time_of_day': [
        'after_midnight', 'morning',
        'afternoon', 'evening', 'night'
    ],
     'city_type':['semi-urban','urban','metropolitian']
}

# Nominal
nominal_cols = [
    'weather',
    'type_of_order',
    'type_of_vehicle',
    'festival'
]
```

```python
df_encoded = df.copy()

df_encoded = binary_encode(df_encoded, binary_map)
df_encoded = ordinal_encode(df_encoded, ordinal_map)
df_encoded = nominal_encode(df_encoded, nominal_cols, drop_first=True)
```
